```python
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
salaries_data = "data/salaries.csv"
titles_data = "data/titles.csv"
```

```python
salaries_df = pd.read_csv(salaries_data)
salaries_df
```

```python
titles_df = pd.read_csv(titles_data)
titles_df
```

```python
merged_tables = pd.merge(salaries_df, titles_df, on = "emp_no", how = "inner")
merged_tables
```

```python
grouped_titles = merged_tables.groupby("title").mean()
grouped_titles
```

```python
titles_salaries_df = grouped_titles.drop(columns = "emp_no")
titles_salaries_df = titles_salaries_df.reset_index()
titles_salaries_df
```

```python
x_axis = titles_salaries_df["title"]
ticks = np.arange(len(x_axis))
y_axis = titles_salaries_df["salary"]
plt.bar(x_axis, y_axis, align="center", alpha=1.0, color=["b", "r", "orange", "g", "y", "gray", "r"])

plt.xticks(ticks, x_axis, rotation="vertical")

plt.ylabel("Salaries ($)")
plt.xlabel("Titles")
plt.title("Average Salary by Title")

plt.savefig("data/Average_salary_by_title.png")

plt.show()
```
